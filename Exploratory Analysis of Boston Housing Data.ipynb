```python
import pandas as pd 
from pandas import DataFrame, Series
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib
from sklearn.datasets import load_boston
pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
print(boston.DESCR)
```

```python
# Convert the numpy arrays into pandas DataFrame using feature names for the column headers 
df = pd.DataFrame(boston.data)
df.columns = boston.feature_names
df.head()
```

```python
# Add the target variable to the pandas DataFrame 
df['MEDV'] = boston.target
df.head()
```

```python
df.describe()
```

```python
# Explore the target variable (MEDV) in greater detail w/ summary statistics using 'describe'. 

df.MEDV.describe()
```

```python
# Next create a histogram of the target variable (MEDV) for a visual of the summary statistics 
import seaborn as sns
sns.distplot(df.MEDV)

from scipy.stats import skew
from scipy.stats import kurtosis

print("Skewness of MEDV: %f" % df.MEDV.skew())
print("Kurtosis of MEDV: %f" % df.MEDV.kurt())
```

```python
# With a skewness a little over 1, the data is slightly positive skewed
# With a kurtosis of 1.5 or so, the data has excess kurtosis or fat tails 
import seaborn as sns

# Per Capita Crime Rate vs. Median Home Value in Boston 
plt.scatter(x = "LSTAT", y = "MEDV", s = 5, data = df, color = 'blue', alpha = 0.6)
```

```python
# Number of Rooms Per Dwelling vs. Median Home Value in Boston 
plt.scatter(x = "RM", y = "MEDV", s = 5, data = df, color = 'green', alpha = 0.6)
```

```python
#box plot Median Home Value vs. Proximity to Charles River 
ind = 'RAD'
data = pd.concat([df['MEDV'], df[ind]], axis=1)
f, ax = plt.subplots(figsize=(10, 4))
fig = sns.boxplot(x=ind, y="MEDV", data=df)
fig.axis(ymin=0, ymax=60);
```

```python
# Concetration of Pollutants vs. Proximity to Working Centers of Boston
ind_2 = 'NOX'
data = pd.concat([df['DIS'], df[ind_2]], axis=1)
f, ax = plt.subplots(figsize=(20, 6))
fig = sns.boxplot(x=ind_2, y="DIS", data=df)
fig.axis(ymin=0, ymax=13);

# Clearly the closer one lives to a working center, the lower the concentration of pollutants are due to less driving.
```

```python
# Correlation Matrix
cmatrix = df.corr()
f, ax = plt.subplots(figsize=(12, 6))
sns.heatmap(cmatrix, vmax = 0.8, cbar=True, annot=True, square=True, cmap = "Blues", fmt='.2f', annot_kws={'size': 8});
```

```python
# Plot a new correlation matrix for the most significant variables 
k = 5 #number of variables for heatmap
cols = cmatrix.nlargest(k, 'MEDV')['MEDV'].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 12}, cmap = "Blues", yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
# It looks like the number of rooms, in addition to distance and zoning are the most significant variables. 
# Create a series of scatterplots to illustrate the relationships between these prediction variables and response variable

sns.set()
cols = ['MEDV', 'DIS', 'B', 'RM']
sns.pairplot(df[cols], size = 3, palette='Set2')
plt.show();
```

```python
# Remove outliers that are more than 3 standard deviations from the mean 
df = df[np.abs(df["MEDV"]-df["MEDV"].mean())<=(2*df["MEDV"].std())]
df.head()
```
